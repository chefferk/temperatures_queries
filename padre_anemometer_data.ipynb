```python
%matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('ggplot')
import os
```

```python
# ---------- first attempt ---------- #
df1 = pd.read_csv('LOGGER00.csv')

# convert time to pandas datetime and sort
df1['month/day/year HH:MM:SS'] = pd.to_datetime(df1['month/day/year HH:MM:SS'])
df1 = df1.sort_values(by='month/day/year HH:MM:SS', ascending=True)

# apply conversion from volts to mph
df1['anemometer 1'] = 0.5949 * df1['anemometer 1'] + 1.1912
df1['anemometer 2'] = 0.5669 * df1['anemometer 2'] + 1.0703
df1['anemometer 3'] = 0.5669 * df1['anemometer 3'] + 1.0703
df1['anemometer 4'] = 0.5717 * df1['anemometer 4'] + 1.1766

# ---------- second attempt ---------- #
df2 = pd.read_csv('LOGGER01.csv')

# convert time to pandas datetime and sort
df2['month/day/year HH:MM:SS'] = pd.to_datetime(df2['month/day/year HH:MM:SS'])
df2 = df2.sort_values(by='month/day/year HH:MM:SS', ascending=True)

# apply conversion from volts to mph
df2['anemometer 1'] = 0.5949 * df2['anemometer 1'] + 1.1912
df2['anemometer 2'] = 0.5669 * df2['anemometer 2'] + 1.0703
df2['anemometer 3'] = 0.5669 * df2['anemometer 3'] + 1.0703
df2['anemometer 4'] = 0.5717 * df2['anemometer 4'] + 1.1766
```
